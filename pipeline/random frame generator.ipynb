```python
from datetime import datetime, timedelta
import pathlib
import random


def generate_random_timestamps(path_str: str, num_timestamps: int) -> list[float]:
    # 1. Zet de string om naar een Path object om makkelijk onderdelen te pakken
    path = pathlib.Path(path_str)

    # 2. Haal de datum- en tijd-onderdelen uit het pad
    parts = path.parts
    date_str = parts[-5]  # "2026_05_22"
    time_str = parts[-3]  # "10_40_00"

    # 3. Combineer ze en zet ze om naar een datetime-object
    datetime_str = f"{date_str} {time_str}"
    start_time = datetime.strptime(datetime_str, "%Y_%m_%d %H_%M_%S")

    # De data loopt 5 minuten, dus de maximale extra tijd is 5 * 60 = 300 seconden
    max_duration_seconds = 5 * 60

    random_timestamps = []

    for _ in range(num_timestamps):
        # 4. Genereer een random aantal seconden
        random_seconds = random.uniform(0, max_duration_seconds)

        # 5. Tel dit op bij de starttijd
        random_time = start_time + timedelta(seconds=random_seconds)

        # 6. Converteer naar Unix timestamp (float)
        epoch_timestamp = random_time.timestamp()

        # We voegen 3 extra random cijfers toe voor de pure nanoseconden
        extra_nanos = random.randint(0, 999) / 1_000_000_000

        # Tel de nanoseconden op bij de timestamp en voeg toe als float
        final_timestamp = epoch_timestamp + extra_nanos
        random_timestamps.append(final_timestamp)

    # Sorteer ze chronologisch
    random_timestamps.sort()

    return random_timestamps


# --- HIER WORDT DE LIST GEDEFINIEERD ---
file_path = r"D:\Data_gathered\2026_05_22\Rosbag\10_40_00\rosbag\rosbag_0.mcap"
aantal_timestamps = 20

# Dit maakt nu een échte Python list aan genaamd TIMESTAMPS
TIMESTAMPS = generate_random_timestamps(file_path, aantal_timestamps)

# --- JE KUNT NU DIRECT MET DE 'TIMESTAMPS' LIST WERKEN ---
print("De list 'TIMESTAMPS' bevat nu de volgende floats:")
print(TIMESTAMPS)
```
